# **5. ì •ë ¬**

NumPyì˜ np.sort() í•¨ìˆ˜ëŠ” ë°°ì—´ì„ ì •ë ¬í•˜ëŠ” ë° ì‚¬ìš©ë©ë‹ˆë‹¤. ì´ í•¨ìˆ˜ëŠ” ê¸°ë³¸ì ìœ¼ë¡œ ì›ë˜ ë°°ì—´ì„ ë³€ê²½í•˜ì§€ ì•Šê³  ì •ë ¬ëœ ë°°ì—´ì˜ ë³µì‚¬ë³¸ì„ ë°˜í™˜í•©ë‹ˆë‹¤.
