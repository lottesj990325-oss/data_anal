# **2. 네이버 웹툰 리뷰 크롤링**

### XPATH
HTML, XML 문서 안에서 특정 요소를 정확하게 찾아가기 위한 경로 표현식

```python
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get('https://comic.naver.com/webtoon/detail?titleId=823195&no=172&week=wed')
time.sleep(2)

print("***베스트 댓글***")
# /html/body/div[1]/div[5]/div/div/div[4]/div[1]/div[3]/div/section/ul/li[1]
# /html/body/div[1]/div[5]/div/div/div[4]/div[1]/div[3]/div/section/ul/li[2]
# /html/body/div[1]/div[5]/div/div/div[4]/div[1]/div[3]/div/section/ul/li[3]
xpath = '/html/body/div[1]/div[5]/div/div/div[4]/div[1]/div[3]/div/section/ul/li'
best_comment_elements = driver.find_elements(By.XPATH, xpath)

for li in best_comment_elements:
    try:
        # //*[@id="wcc_root"]/section/ul/li[1]/div[1]/div[2]/div/p
        # //*[@id="wcc_root"]/section/ul/li[2]/div[1]/div[2]/div/p
        comment_p = li.find_element(By.XPATH, "./div[1]/div[2]/div/p")
        comment_text = comment_p.text.strip()
        print(comment_text)
        print("-" * 30)
    except Exception as e:
        print(e)
```

# **3. 픽사베이 이미지 크롤링**

```python
from urllib.request import Request, urlopen
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
url = "https://pixabay.com/ko/images/search/%ea%b0%95%ec%95%84%ec%a7%80/"
driver.get(url)

# /html/body/div[1]/div[1]/div/div[2]/div[3]/div/div/div/div[1]/div/a/img
image_xpath = '/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div/div/div[1]/div/a/img'

wait = WebDriverWait(driver, 10) # 최대 10초까지 기다림
image_element = wait.until(EC.presence_of_element_located((By.XPATH, image_xpath)))

image_element
 # image_url = image_element.get_attribute('src') 해상도로 인해서 에러?
image_url = image_element.get_attribute('src')
print(image_url) 

image_byte = Request(image_url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/142.0.0.0 Safari/537.36'})
image_byte

with open('dog.jpg','wb') as f:
    f.write(urlopen(image_byte).read())
print('다운로드 완료!!')

driver.quit()
```

# **4. 야놀자 리뷰 크롤링**

```python
def crawl_yanolja_reviews(name, url, max_scroll=100):
    review_list = []
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(3)
    last_height = driver.execute_script("return document.body.scrollHeight")
    # print(last_height)
    same_count = 0
    scroll_times = 0
    while True:
        scroll_times += 1
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            same_count += 1
        else:
            same_count = 0
            last_height = new_height
        if same_count >= 3:
            print('더 이상 새로운 리뷰가 없습니다.')
            break
        if scroll_times >= max_scroll:
            print('최대 스크롤 횟수 도달. 종료합니다.')
            break
    # /html/body/div[1]/section/div/div[5]/div[1]
    # /html/body/div[1]/section/div/div[5]/div[2]
    containers = driver.find_elements(By.XPATH, "/html/body/div[1]/section/div/div[5]/div")
    # print('총 리뷰 개수: ', len(containers))
    for c in containers:
        try:
            review_text = c.text.strip()
        except:
            review_text = ""
        try:
            # /html/body/div[1]/section/div/div[5]/div[2]/div[2]/div/div[1]/div[1]/div[2]/p
            # /html/body/div[1]/section/div/div[5]/div[3]/div[2]/div/div[1]/div[1]/div[2]/p
            date_text = c.find_element(By.XPATH, ".//div/div/div/div/p").text
        except:
            date_text = ""
        review_list.append({
            "review": review_text,
            "date": date_text
        })
    driver.quit()
    return review_list
    
            
```
