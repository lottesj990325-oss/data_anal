````python
import requests
import pandas as pd

def fetch_bike_data():
    base_url = "http://openapi.seoul.go.kr:8088/62486256776c6f74353072744a4779/json/bikeList/"
    start = 1
    end = 1000
    step = 1000
    data_frames = []

    while True:
        # ì˜ˆ: "http://openapi.seoul.go.kr:8088/.../json/bikeList/1/1000"
        url = f"{base_url}{start}/{end}/"
        response = requests.get(url)

        if response.status_code != 200:
            print(f"Status Code: {response.status_code}")
            break

        json_data = response.json()

        try:
            rent_bike_status = json_data["rentBikeStatus"]
            result_code = rent_bike_status["RESULT"]["CODE"]
        except KeyError:
            print("JSON ì˜¤ë¥˜")
            break

        if result_code == "INFO-200":
            print("ë°ì´í„° ì—†ìŒ")
            break

        elif result_code == "INFO-000":
            print(f"ì‹œì‘: {start} ë: {end}")
            try:
                bike_data = rent_bike_status["row"]
                if bike_data:
                    df = pd.DataFrame(bike_data)
                    data_frames.append(df)
            except KeyError:
                print("ë°ì´í„°ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŒ")

        elif result_code == "ERROR-336":
            print("ë°ì´í„°ìš”ì²­ì€ í•œë²ˆì— ìµœëŒ€ 1,000ê±´ì„ ë„˜ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤")
            break

        else:
            print(f"result code: {result_code}")
            break

        start += step
        end += step

    # âœ… while ëë‚œ ë’¤ì— í•œ ë²ˆë§Œ ìµœì¢… ë°˜í™˜
    if data_frames:
        final_df = pd.concat(data_frames, ignore_index=True)
        return final_df
    else:
        return pd.DataFrame()
````

````python
bike_data_df = fetch_bike_data()
````
